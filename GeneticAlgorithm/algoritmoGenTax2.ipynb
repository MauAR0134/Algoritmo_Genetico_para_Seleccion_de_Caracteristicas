```python


# ------------------------------
# Definir genes disponibles
# ------------------------------
LEGO = list(range(len(nombres_columnas2)))  # ajusta según tu dataset

# ------------------------------
# Ejecutar el Algoritmo Genético
# ------------------------------

df_historial, mejoresXgen, mejor_global,best_fitness, parametros = ag.algoritmo_genetico(
    fitness_func=ag.fitness,
    modo=0,
    LEGO=LEGO,
    IND_LEN=10,
    POP_SIZE=10,
    POP_NUM=2,
    GENS=30,
    MUT_RATE=0.1,
    MUT_NULL_RATE=0.1,
    ELITISM=0.1,
    df_expresion=data2,
    target=target2,
    nombres_columnas=nombres_columnas2, 
    modelo="rf",
    metric="accuracy",
    random_state=42
)

# ------------------------------
# Visualización de resultados
# ------------------------------
ag.plot_ag_results(df_historial, mejor_global,best_fitness, parametros,nombres_columnas=nombres_columnas2)
```
